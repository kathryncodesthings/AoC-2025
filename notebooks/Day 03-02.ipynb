```python
test = "987654321111111\n811111111111119\n234234234234278\n818181911112111"

test = test.split()

a = 0
position = 0
big_number = ""

print(test)

for nums in test:
    first_four_characters = nums[:4]
    
    for s in first_four_characters:
        s = int(s)
        if s > a:
            a = s
            position = first_four_characters.index(str(a))
    print(f"{first_four_characters}, largest number is {a} at index {position}")
    big_number += str(a)
    a = 0
    position += 1
    
    second_four_characters = nums[position:5]
    
    for s in second_four_characters:
        s = int(s)
        if s > a:
            a = s
            position = second_four_characters.index(str(a))
    print(f"{second_four_characters}, largest number is {a} at index {position}")
    big_number += str(a)
    a = 0
    
    position = 0
    print (f"Largest number that can be created from first 5 characters of {nums} is {big_number} ")
    big_number = ""
    print ("---")
    
    
```
