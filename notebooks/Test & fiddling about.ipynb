```python
#lists

mylist = ['a', 'b', 'fish', 123, 'Ares', 8, False]

mylist = [1,2,7,3,7,9,3,2,6]

#sort list

mylist.sort(reverse = True)

#remove an item at the specified index

mylist.pop(5)

# count the number of times an object appears in a list

print(mylist.count(7))

print(mylist)

# delete everything from the list

mylist.clear()

mylist
```

```python
# Dictionaries
# create a dictionary
d = {'Key 1':'Value 1', 'Key 2':'Value 2'}

#extract values only to a list (not keys)
newlist = list(d.values())

print (newlist)

newlist = list(d.keys())

#extract keys only to a list (not values)
print (newlist)

# add new key:value pair
d['Colours'] = 'Red'

print(d)

# Create a dictionary that holds lists, tuples and other dictionaries 

blog_post = {
'subject': 'How to live better',
'user': {'name':'dave smith', 'email':'dave@email.com'},
'tags': ('lifestyle', 'religion'),
'likes':['Joe53ph', 'sara77', 'Cathy']}

# add new value to existing list [which is a value in the dictionary]
blog_post['likes'].append('pink_sloth')

print(blog_post)
```

```python
# Sets

# create a set
friends = {"Eleanor", "Chidi", "Tahani","Jason","Michael","Janet","Trevor"}

print(friends)

#delete item from list

friends.discard("Trevor")

print(friends)
```
